```python
import time
import random

# Merge Sort Functions
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = 0
    j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    while i < len(left):
        result.append(left[i])
        i += 1
    while j < len(right):
        result.append(right[j])
        j += 1

    return result

# Binary Search Function
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif target < arr[mid]:
            high = mid - 1
        else:
            low = mid + 1
    return -1

#  Input Sample
delivery_points = [52, 13, 76, 24, 5, 89, 31, 47, 66, 18, 90, 7]
print("Original Delivery Points:", delivery_points)

#  Sorting
sorted_list = merge_sort(delivery_points)
print("Sorted Delivery Points:", sorted_list)
```

```python
target = 47
index = binary_search(sorted_list, target)
if index != -1:
    print(f"Searching for {target} → FOUND at index {index}")
else:
    print(f"Searching for {target} → NOT FOUND")
```

```python
size1 = 50
size2 = 500

arr1 = [random.randint(1, 5000) for _ in range(size1)]
arr2 = [random.randint(1, 10000) for _ in range(size2)]

# Time for size1
start = time.time()
merge_sort(arr1)
time1 = (time.time() - start) * 1000  # in ms

# Time for size2
start = time.time()
merge_sort(arr2)
time2 = (time.time() - start) * 1000  # in ms

print(" Execution times for the two different input sizes.")
print("Time for", size1, "items:", time1, "ms")
print("Time for", size2, "items:", time2, "ms")
```
